# function 

###  함수를 만드는데 숫자, 계산할 숫자, 계산 부호를 넣어서 함수를 만든다.
## +,-,*,/
## 계산부호가 잘못 들어왔을 경우에는 잘못된 입력 False를 리턴

```python
def calc(num1,num2,cal):
    if cal == '+':
        return num1+num2
    elif cal == '-':
        return num1-num2
    elif cal == '*':
        return num1*num2
    elif cal == '/':
        return num1/num2
    else:
        print('잘못된 입력입니다')
        return False
```

```python
calc(0.1,0.1,'+')
```
